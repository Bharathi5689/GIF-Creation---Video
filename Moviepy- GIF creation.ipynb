```python
from IPython.display import Video

# Display the input video
video_path = "./data/dog_video.mp4"
Video(video_path)
```

```python
from moviepy.editor import *

# Load the video clip
clip = VideoFileClip(video_path)

# Trim the video to the first 4 seconds
clip = clip.subclip(0,4)

# Rotate the video by 360 degrees
clip = clip.rotate(360)

# Create a text clip with the message "Cuteone"
txt_clip = TextClip("Cuteone",fontsize = 100, color ="red")

# Set the position of the text clip to the center and its duration to 3 seconds
txt_clip = txt_clip.set_pos("center").set_duration(3)

# Combine the original video clip and the text clip
gif = CompositeVideoClip([clip,txt_clip])

# Write the resulting GIF file with a frame rate of 240 frames per second
gif.write_gif("Doggy.gif", fps = 240)

# Display the resulting GIF in the notebook with a width of 240 pixels
gif.ipython_display(width = 240)
```
